# Loading USArrest data

```python
# loading data
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("USArrests.csv")
df.shape
# df_raw.head()

# plots = 20
# df = df_raw.iloc[:, 0:plots]

display(df.describe())
display(df.head())
display(df.columns)
```

# Estandarizando los datos

```python
# Data preprocessing
from sklearn.preprocessing import scale
import numpy as np

# Center in 0
scaled_df = scale(df.iloc[:, 1:])
print("scaled_df shape: ", scaled_df.shape)
# plot distribution
fig, ax = plt.subplots( 1, 4, tight_layout = True, figsize = (20, 4))
for i in range(4):
    
    ax[i].hist(scaled_df[:, i])
```

Desde histogramas, se observa que todas las variables estan centradas en 0, lo que valida visualmente el escalado de los datos.

# Resolviendo PCA manualmente

```python
from numpy import linalg

# getting cov matrix
# cov = np.cov(scaled_df)
C = np.dot(scaled_df.T, scaled_df) / (50-1)

# print(C.shape)
# print("cov matrix: \n", C)
# getting eigen values and vectors
eig_values, eig_vectors = linalg.eig(C)
print("eig vector: \n", eig_vectors)
print("eig values: \n", eig_values)
# Project X onto PC space
df_pca = np.dot(scaled_df, eig_vectors)
# print(df_pca)
```

```python
# Plot PCs
fig, ax = plt.subplots(figsize = (10,5))
ax.scatter(df_pca[:,0], df_pca[:, 1], label = df.iloc[:, 0])
for i, txt in enumerate(df.iloc[:, 0]):
#     print(i)
    ax.annotate(txt, (df_pca[i,0], df_pca[i,1]))
```

```python
# Computing PVE
PVEs = eig_values/np.sum(eig_values)
print("Explained variance: ", PVEs)
print("Explained variance with first 2 PC: ", PVEs[0] + PVEs[1], " %")

fig, ax = plt.subplots(2, 1)
ax[0].bar(np.arange(0, 4), PVEs)
ax[1].bar(np.arange(0, 4), np.cumsum(PVEs))
```

Desde gráficos de proporcion de varianza explicada, se tiene que PC 1 explica el 62 [%] de varianza de los datos. Luego PC 2 explica 25 [%] de varianza. Por lo tanto juntos, PC 1 y PC 2, explican 87 [%] de variabilidad de los datos originales.

# PCA utilizando funciones pre-contruidas de python

```python
# loading data
import pandas as pd
import matplotlib.pyplot as plt

df_raw = pd.read_csv("USArrests.csv")
df_raw.shape
```

```python
# Standarizing data
from sklearn.preprocessing import StandardScaler

df = df_raw.iloc[:, 1:]
df = StandardScaler().fit_transform(df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(df)

# principalComponents.shape

# display(df_raw.iloc[:, 0])
# converting to DF
pca_df = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
# pca_df = pd.concat([df_raw.iloc[:, 0], pca_df], ignore_index=True)
pca_df["state"] = df_raw.iloc[:, 0]
pca_df.head()
```

Dataframe contiene las coordenadas en los componentes principales de cada observación.

```python
# plotting explained variance
fig, ax = plt.subplots(2, 1, tight_layout = True)
ax[0].bar(np.arange(2), pca.explained_variance_ratio_)
ax[0].set_title("Explained variance by PC")
ax[1].bar(np.arange(2), np.cumsum(pca.explained_variance_ratio_))
ax[1].set_title("Explained cumulative variance by PC")
print(pca.explained_variance_ratio_)
```

Desde gráfico de varianza expicada por PCs, se tiene que PC 1 y PC 2 juntos, explican el 87 [%] de la variabilidad de los datos originales. Resultados similares se obtuvieron con procedimiento hecho manualmente.

```python
# plotting data projection with PCA
fig, ax  = plt.subplots(figsize = (10,7))
ax.scatter(pca_df["principal component 1"], pca_df["principal component 2"])

for i, txt in enumerate(pca_df["state"]):
#     print(txt)
    ax.annotate(txt, (pca_df.iloc[i, 0], pca_df.iloc[i,1]))
    
```

Se observa que se tienen representaciones similares respecto a los resultados obtenidos manualmente, ya que los estados se distribuyen de forma similares respecto al otro procedimiento manual.
